# REGULAR EXPRESSION

```python
import re
text = "foo    bar\t baz  \tqux"
re.split('\s+',text)# '\s' space,newline and tabe and '+' means all
```

```python
regex=re.compile('\s+')
```

```python
regex.split(text)
```

```python
regex.findall(text)
```

```python
tet = "Dave dave@google.comSteve steve@gmail.comRob rob@gmail.comRyan ryan@yahoo.com"
pattern = r'[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}'
```

```python
rg = re.compile(pattern, flags=re.IGNORECASE)
rg.findall(tet)
```

```python
nc='this is my nic number 42401-6755778-5'
```

```python
nicpattern=r'[0-9]{5}\-[0-9]{7}\-[0-9]{1}'
```

```python
rg = re.compile(nicpattern)
rg.findall(nc)
```
